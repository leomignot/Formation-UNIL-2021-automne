## Les principaux types de Python

Connaître le type d'un objet : `type(objet)`

### Le nom des variables

Tous les caractères sauf le `.`, les espaces et les mots clés de Python. Nommer de manière explicite.

### Les entiers (*int*)

``` python
un_entier = 10
```

### Les nombres à virgules (*float*)

- arrondir un nombre à trois décimales avec la fonction `round(nombre, 3)`

### Le binaire (*bool*)

Deux valeurs possibles `True` (vraie) et `False` (fausse)

### Les chaînes de caractères (*str*)

``` python
texte = "Ceci est une chaîne de caractères"
``` 

- Plusieurs manières de définir une chaîne de caractères en termes de guillemets : `",',"""` 
- Une chaîne de caractère est définie par un encodage
- Attention aux caractères spéciaux : possible de les échapper avec `\`
- Sélection d'une partie de la chaîne avec `texte[2:10]`
- Des méthodes utiles : mettre en majuscule `texte.upper()`, en minuscule `texte.lower()`, couper quand un motif est rencontré `texte.split(motif)`, remplacer un motif1 par un motif2 `texte.replace(motif1,motif2)`
- Longueur d'une chaîne `len(texte)`
- Concaténer deux chaînes : `"Ceci" + " est un texte"`

### Les listes (*list*)

Une liste est un ensemble défini par des crochets, indexée par les entiers, et modifiable

``` python
liste = [1,"deux",3.0]
``` 

- Accéder à l'élément à la position X (**qui va de 0** à au nombre d'éléments-1) `liste[X]`
- Dernier élément de la liste `liste[-1]`
- Tous les éléments entre le premier et le troisième `liste[0:3]`
- Index d'un élément : `liste.index("deux")`
- Ajouter un élément `liste.append("quatre")` ; supprimer un élément `liste.remove(1)`
- Trier une liste `liste.sort()`
- Joindre les éléments d'une liste avec un élément, ici le tiret : `"-".join(liste)`

### Les dictionnaires (*dict*)

Un dictionnaire est un ensemble avec chaque élément qui est défini comme un couple `clé:valeur`

``` python
dictionnaire = {"un":"première entrée","2":2}
``` 

- Une valeur peut être retrouvée par sa clé : `dictionnaire["un"]`
- Un élément peut être rajouté directement : `dictionnaire["trois"] = "Nouvelle entrée"`
- Les clés peuvent être récupérées par `dictionnaire.keys()`


### *Set* et *tuples*

Deux types de contenant :
- set : ensemble non ordonné contenant une version unique de chaque élément `un_set = {"Un","Deux"}
- tuple : liste non modifiable `un_tuple = ("Un","Deux")`

### Le type vide

`None`


## Transformer les types

Chaque type a un constructeur associé qui permet (sous certaines conditions) de changer de type.

Par exemple transformer un nombre en chaîne de caractères : `str(10)` ou `list("texte")`


## Ressources 

- Une version condensée en anglais : https://perso.limsi.fr/pointal/_media/python:cours:mementopython3-english.pdf
