## Les principaux mots-clés de Python

### Les signes utilisés

- `=` allocation d'une variable
- `[]` définir une liste ou sélectionner un élément
- `{}` définir un dictionnaire
- `#` pour commencer une ligne

### Evaluer une condition

évaluer une condition revient à vérifier si quelque chose est vrai `True` ou faux `False`. Pour cela, il y a des opérateurs de comparaison : `>`, `<`, `<=`, `>=`, `==`, `!=`, ou encore `in`.

Ainsi, pour savoir si un élément est dans une liste, on peut écrire la condition `element in liste` qui prendra la valeur vraie ou fausse suivant la valeur de l'élément et le contenu de la liste


### Les fonctions de base

- `print(element)` afficher l'élément
- `type(objet)` type de l'objet
- `len(ensemble)` nombre d'éléments dans l'ensemble
- `str(element)` transforme un élément en chaîne de caractères (si c'est possible) ; `list(element)` pareil en liste ; `int(element)` en entier, etc.
- `round(nombre,decimales)` arrondir un nombre pour ne garder que le nombre de décimales indiquées
- `range(x,y)` générer la liste des entiers entre x et y-1

Les autres fonctions de base du langage : https://docs.python.org/fr/3.6/library/functions.html

### Définir des actions conditionnelles

Pour tester une condition avant d'exécuter un morceau de code, on utilise le mot clé `if`, et éventuellement (mais pas nécessaire) `else` pour mettre du code exécuté si la condition est fausse

``` python
if CONDITION:
    code exécuté si vrai
else:
    code exécuté si faux
```

S'il y a différentes conditions, il est possible d'utiliser le mot clé `elif`


### Faire une boucle

Pour prendre un à un les éléments d'un ensemble (par exemple d'une liste) on peut faire un boucle avec le mot clé `for` et une variable intermédiaire (qui est crée pour la boucle) auquel on donne le nom que l'on souhaite


``` python
for element in ensemble:
    instructions ...
```

(une autre manière de faire une boucle est d'utiliser la structure `while CONDITION:` (tant que ...)


### Définir une fonction

Pour définir une fonction, on définit ce qui doit être mis en entrée (les arguments), les étapes que la fonction doit réaliser, et ce qu'elle doit retourner une fois les opérations réalisées. Une fois la fonction définie, on peut l'utiliser dans des cas particulier.

``` python 
def nom_de_la_fonction(arguments):
    opérations ...
    return elements_a_retourner
```

Une fois définie on peut utiliser la fonction `nom_de_la_fonction(arguments_particuliers)`

### Ouvrir un fichier

``` python
with open(nomdufichier,mode) as file:
    opérations sur le fichier ....
```

- le mode peut être lecture `"r"` ou écriture `"w"`
- le nom du fichier est l'ensemble du chemin vers le fichier à partir du répertoire courant
- les opérations peuvent alors être par exemple `file.write(contenu)` (en mode écriture) ou `file.read()` (en mode lecture). D'autres fonctions existent.
