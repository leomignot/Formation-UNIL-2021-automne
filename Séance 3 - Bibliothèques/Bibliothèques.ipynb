## Exercice : faire un script qui vérifie si le mot hydroxychloroquine est présent sur les sites d'informations : 
- https://www.heidi.news/
- https://www.letemps.ch/

Décomposer les opérations :

0. Définir le mot clé
1. Récupérer la page d'un site -> requests
2. Récupérer uniquement le texte -> passage en chaîne de caractères
3. Tester la présence du mot
4. Afficher le résultat


```python
# 0. Définir les variables
url = "https://www.heidi.news/"
motcle = "chloroquin"

# 1. Récupérer la page avec Requests
page = requests.get(url)

# 2. Récupération du texte
contenu = page.text

# 3. Tester la présence d'un mot
if motcle in contenu:
    affichage = "Le mot "+motcle+" est dans la page "+url
else:
    affichage = "Le mot "+motcle+" n'est pas dans la page "+url
    
# 4. Afficher le résultat
print(affichage)
```

On va généraliser : créons une fonction qui prend en entrée une url et un mot clé, et renvoie Vrai ou Faux suivant la présence du mot

```python
def presence_mot(url,motcle):
    """
    Fonction de présence d'un mot clé motcle sur une page à l'adresse url
    """
    page = requests.get(url)
    if motcle in page.text:
        return "Le mot "+motcle+" est dans la page "+url
    else:
        return "Le mot "+motcle+" n'est pas dans la page "+url
```

```python
presence_mot("https://www.heidi.news","chloroquin")
```

On généraliste sur tous nos liens

- https://www.heidi.news/
- https://www.letemps.ch/
- https://www.lemonde.fr/

```python
# Définir une fonction
def presence_mot(url,motcle):
    """
    Fonction de présence d'un mot clé motcle sur une page à l'adresse url
    """
    page = requests.get(url)
    if motcle in page.text:
        return "Le mot "+motcle+" est dans la page "+url
    else:
        return "Le mot "+motcle+" n'est pas dans la page "+url

# Définir une liste
liste_urls = ["https://www.heidi.news/",
               "https://www.letemps.ch/",
            "https://www.lemonde.fr/"]

# Faire une boucle sur la liste
for u in liste_urls:
    resultat = presence_mot(u,"chloroquin")
    print(resultat)
```

La fonction n'est pas nécessaire

```python
# 0. Définir les variables
liste_urls = ["https://www.heidi.news/",
              "https://www.letemps.ch/",
              "https://www.lemonde.fr/"]
motcle = "chloroquin"

# Boucle sur l'ensemble des urls
for url in liste_urls:
    
    # 1. Récupérer la page avec Requests
    page = requests.get(url)

    # 2. Récupération du texte
    contenu = page.text

    # 3. Tester la présence d'un mot
    if motcle in contenu:
        affichage = "Le mot "+motcle+" est dans la page "+url
    else:
        affichage = "Le mot "+motcle+" n'est pas dans la page "+url

    # 4. Afficher le résultat
    print(affichage)
```

## Aparte sur les fonctions

return est différent de print

```python
# Retourner hors de la fonction
def addition(x):
    return x+1
```

```python
addition(10)
print(addition(100))
```

```python
# Afficher à l'écran et renvoyer rien
def afficher_addition(x):
    print(x+1)
    return None
```

```python
afficher_addition(10) + 1
```
